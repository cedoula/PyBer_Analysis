```python
%matplotlib inline
```

```python
# Import depedencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot().

# Create the figure
fig, ax = plt.subplots()
# Define the axes
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot().
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot().
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label="Boston")
# Create label for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker= "*", color="blue", linewidth=2, label="Boston")
# Create label for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create plot with object oriented approach
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0, 45)
ax.grid()
ax.legend()
```

```python
# Create the bar chart.
plt.bar(x_axis, y_axis)
```

```python
# Create the bar chart.
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fares($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the horizontal bar chart.
plt.barh(x_axis, y_axis)
```

```python
# Create the horizontal bar chart.
plt.barh(y_axis, x_axis)
```

```python
# Create the horizontal bar chart.
plt.barh(x_axis, y_axis)
# Invert y axis
plt.gca().invert_yaxis()
```

```python
# Create the horizontal bar chart.
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
# Invert y axis.
plt.gca().invert_yaxis()
# Add title
plt.title("PyBer Fare by Month")
# Add axes labels
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Add legend.
plt.legend()
```

```python
# Create the bar chart with ax.bar()
ax = plt.axes()
ax.bar(x_axis, y_axis)
```

```python
# Create the bar chart with ax.bar()
ax = plt.axes()
ax.bar(x_axis, y_axis, color="k", label="New York")
ax.legend()
```

```python
# Create the bar chart with ax.bar()
ax = plt.axes()
ax.barh(x_axis, y_axis)
```

```python
# Create the bar chart with ax.bar()
ax = plt.axes()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
plt.gca().invert_yaxis()
ax.set_title("PyBer Fares by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.legend()
```
